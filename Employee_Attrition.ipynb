```python
# 3) Drop Unwanted Columns
df.drop(columns=['EmployeeCount','EducationField','EmployeeNumber','Over18','TrainingTimesLastYear','StandardHours','MonthlyRate','DailyRate','HourlyRate'],inplace=True)
```

```python
df.columns
```

```python
# 4) Encoding dataframe
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Attrition'] = le.fit_transform(df['Attrition'])
```

```python
order_col = ['Gender','MaritalStatus','OverTime','BusinessTravel','JobRole']
```

```python
from sklearn.preprocessing import OrdinalEncoder
order = {
    'Gender':['Female', 'Male'],
    'MaritalStatus':['Single', 'Married', 'Divorced'],
    'OverTime':['Yes', 'No'],
    'JobRole': ['Human Resources', 'Laboratory Technician', 'Healthcare Representative', 'Sales Representative', 'Sales Executive', 'Manager', 'Manufacturing Director', 'Research Scientist', 'Research Director'],
    'BusinessTravel': ['Non-Travel', 'Travel_Rarely', 'Travel_Frequently']
}

order_col = list(order.keys())

ordinal_encoder = OrdinalEncoder(categories= [order[col] for col in order_col])
```

```python
nominal_col = ['Department']
```

```python
from sklearn.preprocessing import OneHotEncoder

one_hot_enc = OneHotEncoder(drop='first',handle_unknown='ignore')
```

```python
from sklearn.compose import ColumnTransformer

preproessor = ColumnTransformer([
    ('ordinal',ordinal_encoder,order_col),
    ('nominal',one_hot_enc,nominal_col)
],remainder='passthrough')
```

```python
x = df.drop(columns=['Attrition'])
y = df['Attrition']
```

```python
df_e = preproessor.fit_transform(x)
```

```python
df = pd.DataFrame(df_e,columns=preproessor.get_feature_names_out())
```

```python
df.columns = (
    df.columns
    .str.replace("ordinal","")
    .str.replace("nominal","")
    .str.replace("remainder","") 
    .str.replace("__","")               
)
```

```python
df.head()
```

```python
df=pd.concat([y,df],axis=1)
```

```python
num_cols = df.select_dtypes(include='number').columns
num_cols
```

```python
# 6) Finding Outliers
outlier_summary = {}

for col in num_cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3-Q1
  lower_limit = Q1-1.5*IQR
  upper_limit = Q3+1.5*IQR

  outlier = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
  print(f"{col} has {len(outlier)} outliers")
  outlier_percentage = round(len(outlier)/len(df)*100,2)

  outlier_summary[col] = {
      'Q1':Q1,
      'Q3':Q3,
      'IQR':IQR,
      'lower_limit':lower_limit,
      'upper_limit':upper_limit,
      'outliers count':len(outlier),
      'outlier_percentage %':outlier_percentage
  }
```

```python
outlier_df = pd.DataFrame(outlier_summary).T

outlier_df
```

```python
# 8) Select Features And Target
x = df.drop(columns=['Attrition'])
y = df['Attrition']

print("Shape of X:", x.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 9) Train Test Split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print("Shape of X_train:", x_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_test:", x_test.shape)
print("Shape of y_test:", y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    'Logistic Regression':(LogisticRegression(),{}),
    'KNN':(KNeighborsClassifier(),{'model__n_neighbors':[1,3,6,7,10],'model__weights':['uniform','distance']}),
    'Decision Tree':(DecisionTreeClassifier(),{'model__max_depth':[1,4,6,8,10],'model__criterion':['gini','entropy']}),
    'Random Forest':(RandomForestClassifier(),{'model__n_estimators':[50,100,150,200],'model__criterion':['gini','entropy']})
}
```

```python
from imblearn.pipeline import Pipeline as ImbPipeline
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import *
```

```python
results = {}

for name,(model,params) in models.items():
  pipe = ImbPipeline([
                      ('scaler',StandardScaler()),
                      ('smote',SMOTE()),
                      ('model',model)
  ])

  grid = GridSearchCV(pipe,params,cv=5,scoring='accuracy')
  grid.fit(x_train,y_train)

  best_model = grid.best_estimator_ 
  y_hat = best_model.predict(x_test)

  y_hat = grid.predict(x_test)
  acc = accuracy_score(y_test,y_hat)
  f1 = f1_score(y_test,y_hat)
  recall = recall_score(y_test,y_hat)
  precision = precision_score(y_test,y_hat)

  results[name] = {
      'model':name,
      'best_params':grid.best_params_,
      'score':grid.best_score_,
      'accuracy':acc,
      'f1_score':f1,
      'recall':recall,
      'precision':precision,
  }
 
```

```python
result_df = pd.DataFrame(results).T
result_df
```
